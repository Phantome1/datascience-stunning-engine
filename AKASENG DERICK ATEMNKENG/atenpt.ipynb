```python
## 1ï¸âƒ£ Connexion et insertion des donnÃ©es dans MongoDB
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017/")
db = client["shopcam"]

clients = db["clients"]
produits = db["produits"]
commandes = db["commandes"]

# Exemple dâ€™insertion
clients.insert_many([
    {"_id": 1, "nom": "Tchouameni", "ville": "YaoundÃ©", "sexe": "M", "age": 29, "inscription": "2024-03-10"},
    {"_id": 2, "nom": "Ndongo", "ville": "Douala", "sexe": "F", "age": 24, "inscription": "2024-05-20"}
])
```

```python
## 2ï¸âƒ£ RequÃªtes MongoDB avec PyMongo

#Exemples de commandes :

# Afficher tous les clients
for c in clients.find():
    print(c)

# Clients de YaoundÃ©
for c in clients.find({"ville": "YaoundÃ©"}):
    print(c)

# Nombre total de commandes
print(commandes.count_documents({}))
```

```python
## 3ï¸âƒ£ Connexion MongoDB â†” PySpark
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("MongoDB_ShopCam_Analysis") \
    .config("spark.mongodb.input.uri", "mongodb://127.0.0.1/shopcam.commandes") \
    .config("spark.mongodb.output.uri", "mongodb://127.0.0.1/shopcam.resultats") \
    .getOrCreate()

df_commandes = spark.read.format("mongodb").load()
df_commandes.show(5)
```

```python
#4ï¸âƒ£ Analyse des donnÃ©es avec PySpark
#a) Chiffre dâ€™affaires par produit
df_commandes.groupBy("produit_id").sum("quantite").show()
```

```python
#b) Moyenne dâ€™Ã¢ge des clients par ville
df_clients = spark.read.format("mongodb").option("uri", "mongodb://127.0.0.1/shopcam.clients").load()
df_clients.groupBy("ville").avg("age").show()
```

```python
#c) Statistiques sur les commandes livrÃ©es
df_commandes.filter(df_commandes.statut == "livrÃ©e").count()
```
