```python
from pyspark import SparkContext

# IMPORTANT: Stop the old, mismatched session first
try:
    sc.stop()
except:
    pass

# Start the fresh session (it will now pick up the 3.13 version)
sc = SparkContext("local", "WordCountFix")
```

```python
# Load the input text file

lines = sc.textFile("wordcount.txt")
```

```python
# Split each line into words

words = lines.flatMap(lambda line: line.split(" ")).filter(lambda w: w != "")
```

```python
# Map each word to a tuple of (word, 1)
word_counts = words.map(lambda word: (word, 1))
```

```python
# Reduce by key to get the count of each word
#word_counts = word_counts.reduceByKey(lambda a, b: a + b)

word_pairs = words.map(lambda word: (word, 1))
word_counts = word_pairs.reduceByKey(lambda a, b: a + b)
```

```python
# Collect the results and print them
#results = word_counts.collect()
#for result in results:
 #   print(result)

    # This should now work without the version error
results = word_counts.collect()
for r in results:
    print(r)
```
