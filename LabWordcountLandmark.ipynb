```python
# Create SparkSession 
from pyspark import SparkContext
```

```python
# Check if a Spark context already exists
sc = SparkContext.getOrCreate()

if sc is None:
    # Create a new Spark context
    sc = SparkContext("local", "WordCount")
```

```python
sc
```

```python
# Load the input text file
lines = sc.textFile("wordcount.txt")
```

```python
# Split each line into words
words = lines.flatMap(lambda line: line.split(" "))
```

```python
# Map each word to a tuple of (word, 1)
word_counts = words.map(lambda word: (word, 1))
```

```python
# Reduce by key to get the count of each word
word_counts = word_counts.reduceByKey(lambda a, b: a + b)
```

```python
# Collect the results and print them
results = word_counts.collect()
for result in results:
    print(result)
```

```python
# Stop the Spark context if it was created in this script
if not SparkContext.getOrCreate():
    sc.stop()
```
